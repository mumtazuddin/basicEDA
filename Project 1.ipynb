### DATA CLEANING 

First, we check for how many null values are there in each column.

```python
df.isnull().sum()
```

No null values in this csv file, but there was some problem where it was adding zeros on the **zip** column where there was a 4digit zipcode and 3digit zipcode because most of the zipcode had 5digit codes.  

```python
## Add leading zeros to the integer column in Python data frame 
 
df['zip']=df['zip'].apply(lambda x: '{0:0>5}'.format(x))
print(df.zip)
```
